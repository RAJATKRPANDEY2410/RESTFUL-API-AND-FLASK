```python
#ANSWERS TO THE QUESTIONS

#Question 1: What is a RESTful API?
A RESTful API is an application programming interface that follows the principles of Representational State Transfer (REST). It uses HTTP methods to perform operations on resources, which are represented by URLs. REST emphasizes stateless communication, scalability, and simplicity.


#Question 2: What is Flask, and why is it popular for building APIs?
Flask is a lightweight Python web framework designed for building web applications and APIs. It is popular because:
- It is simple and easy to learn.
- Provides flexibility with minimal boilerplate code.
- Has strong community support and extensions (e.g., Flask-SQLAlchemy).
- Ideal for rapid prototyping and small-to-medium applicatio


#Question 3: What are HTTP methods used in RESTful APIs?
Common HTTP methods include:
- GET → Retrieve data from the server.
- POST → Send new data to the server.
- PUT → Update existing data.
- DELETE → Remove data.
- PATCH → Partially update data.


```

```python
from flask import Flask

#Question 4: What is the purpose of the @app.route() decorator in Flask?

#The @app.route() decorator maps a URL endpoint to a function in Flask. It tells Flask which function should handle requests to a specific path. For example

app = Flask(__name__)

@app.route('/hello')
def hello():
    return "Hello, World!"
```

```python
#Question 5: What is the role of Flask-SQLAlchemy?
Flask-SQLAlchemy is an extension that integrates SQLAlchemy ORM with Flask. It simplifies database operations by:
- Allowing Python classes to represent database tables.
- Managing queries and relationships.
- Handling migrations and connections efficiently



```

```python
#Question 6: How do you create a basic Flask application?

from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return "Welcome to Flask!"

if __name__ == '__main__':
    app.run(debug=True)
```

```python
#Question 7: How do you return JSON responses in Flask?

from flask import jsonify

@app.route('/data')
def data():
    return jsonify({"message": "Hello", "status": "success"})
```

```python
#Question 8: How do you handle POST requests in Flask?

from flask import request

@app.route('/submit', methods=['POST'])
def submit():
    data = request.json
    return jsonify({"received": data})
```

```python
#Question 9: How do you handle errors in Flask (e.g., 404)

@app.errorhandler(404)
def not_found(error):
    return jsonify({"error": "Resource not found"}), 404
```

```python
#Question 10: How do you structure a Flask app using Blueprints

from flask import Blueprint

api = Blueprint('api', __name__)

@api.route('/hello')
def hello():
    return "Hello from Blueprint!"

# Register in main app
app.register_blueprint(api, url_prefix='/api')
```
